### Combining and merging data sets

```python
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df2 = DataFrame({'key': ['a', 'b', 'd'],
                 'data2': range(3)})
df1
```

```python
df2
```

```python
pd.merge(df1,df2) # inner join on key
```

```python
pd.merge(df1, df2, on='key') # specified the joined column, same as above
```

```python
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df4 = DataFrame({'rkey': ['a', 'b', 'd'],
                 'data2': range(3)})
df3
```

```python
df4
```

```python
pd.merge(df3, df4, left_on='lkey',right_on='rkey') # different column name, specified, don't overlap
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer') # outer join
```

```python
df5 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                 'data1': range(6)})
df6 = DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                 'data2': range(5)})
df5
```

```python
df6
```

```python
pd.merge(df5,df6,how='outer')
```

```python
pd.merge(df5,df6,how='inner')
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'rval': [4, 5, 6, 7]})
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
pd.merge(left,right,on='key1')
```

```python
pd.merge(left,right,on='key1',suffixes=('_left','_right'))
```

```python
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                  'value': range(6)})
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
left1
```

```python
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True,how='outer')
```

```python
# hierarchical index
left2 = DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
                   'key2': [2000, 2001, 2002, 2001, 2002],
                   'data': np.arange(5.)})
right2 = DataFrame(np.arange(12).reshape((6, 2)),
                   index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                          [2001, 2000, 2000, 2000, 2001, 2002]],
                   columns=['event1', 'event2'])
left2
```

```python
right2
```

```python
pd.merge(left2, right2, left_on=['key1','key2'], right_index=True)
```

```python
pd.merge(left2, right2, left_on=['key1','key2'], right_index=True, how='outer')
```

```python
left3 = DataFrame([[1., 2.], [3., 4.], [5., 6.]], index=['a', 'c', 'e'],
                 columns=['Ohio', 'Nevada'])
right3 = DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                   index=['b', 'c', 'd', 'e'], columns=['Missouri', 'Alabama'])
left3
```

```python
right3
```

```python
pd.merge(left3,right3,left_index=True,right_index=True) #merge on two side's index
```

```python
pd.merge(left3,right3,left_index=True,right_index=True,how='outer') #merge on two side's index
```

```python
left3.join(right3) #left join
```

```python
left3.join(right3, how='inner') #left join
```

```python
left3.join(right3, how='outer') #DataFrame.join() to merge on index, func same as above
```

```python
left2.join(right2, on=['key1','key2'])  #merge caller and parameter on caller's column(on)  and para's index, default left join
```

```python
left2.join(right2, on=['key1','key2'], how='inner')
```

```python
another = DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                    index=['a', 'c', 'e', 'f'], columns=['New York', 'Oregon'])
another
```

```python
left3.join([right, another]) #default left join
```

### Transforming data using a function or mapping

```python
data = DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami',
                           'corned beef', 'Bacon', 'pastrami', 'honey ham',
                           'nova lox'],
                  'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}

data['animal'] = data['food'].map(str.lower).map(meat_to_animal) # use dictionary object
data
```

```python
data['animal2'] = data['food'].map(lambda x: meat_to_animal[x.lower()])
data
```

### Discretization and bining

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

```python
cats.categories
```

```python
cats.codes
```

```python
pd.value_counts(cats)
```

```python
cats2 = pd.cut(ages, bins, right=False) #[ )
cats2
```

```python
pd.value_counts(cats2)
```

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
cats3 = pd.cut(ages, bins, labels=group_names)
cats3
```

```python
pd.value_counts(cats3)
```

```python
data = np.random.rand(20)
cats4 = pd.cut(data, 4, precision=2) #divide min-max into 4 equal part
cats4
```

```python
pd.value_counts(cats4)
```

```python
data2 = np.random.randn(2000)
cats5 = pd.qcut(data2, 4) #divide all data into 4 equal part
cats5
```
